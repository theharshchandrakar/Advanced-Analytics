```python
import numpy as np
import matplotlib.pyplot as
plt
```

```python
#Investment parameterrs
investment = 10000          #investing Rs1 lakh
num_years = 20              #for 20 years
stock_weight = 0.6          #60% in stocks 
bond_weight = 0.4           #40% in stocks 
```

```python
# Return and standard deviation for stocks and bonds

stock_return = 0.15 # 15% per year, high returns
stock_sd = 0.60 # but with a high variation

bond_return = 0.07 # 7% per year, low returns
bond_sd = 0.05 # but more consistent

# Number of simulations
iterations = 5555
```

#Simulate random returns for stocks and bonds

#np.random.normal: This function generated random returns for stocks and bonds

#for each year of the investment horizon in each simulation

#Arguments: mean (average return), standard deviation (volatility), and simulation size (no of years and simulations)

#stock returns and bond returns will be 5555 x 20 arrays

```python
stock_returns = np.random.normal(stock_return, stock_sd, (iterations, num_years))
bond_returns = np.random.normal(bond_return, bond_sd, (iterations, num_years))

print(stock_returns)
print(bond_returns)
total_results_pct = ((stock_weight * stock_returns) + (bond_weight * bond_returns)) * 100
```

#Supress scientific notation for NumPy arrays

```python
np.set_printoptions(suppress=True)
print(total_results_pct)
```

```python
# Create a 5555 x 20 arrray that will store cumultive investment valeus
portfolio_values = investment * (1+ (total_results_pct / 100)).cumprod(axis=1)
print("***Portfolio Values***")
print(portfolio_values)
```

```python
# Calculate final portfolio value stats
final_values = portfolio_values[:,-1]
print(f"Summary of final portfolio values after {num_years} years:")
print(f'Mean: {final_values.mean():.2f}')
print(f' Standard Deviation: {final_values.std():.2f}')
```

```python
#Calculate percentiles
percentiles = np.percentile(final_values, [25,50,100])
print(f'\nPercentiles of final portfolio values:')
print(f'25th percentile: {percentiles[0]:.2f}')
print(f'50th percentile(median): {percentiles[1]:.2f}')
print(f'100th percentile: {percentiles[2]:.2f}')
```
