Understanding Monte Carlo Simulation

```python
import numpy as np
```

Simulation of rolling a dice twice, min number we get is 1, max is 6

Add the result of the two simulations, so possible values will be:
(1,1) or (1,2)...(6,6)... SUm will be between 2 and 12

```python
#part 1
def roll_dice():
    #return np.random.randint(1,7,2)     #Generate an array of 2 random integers bw 1to6(both inclusive) and add them
    return np.sum(np.random.randint(1,7,2))

print(roll_dice())
```

Someone Approaches us saying i will give u $5 if u get '7' but will take $1 if you get a number other than 7.

How do we know what will happen?


```python
#Part 2
#Our Own Monte Carlo Simulation like function

def monte_carlo_simulation(run=1000):
    results = np.zeros(2)
    for _ in range(run):
        if roll_dice() == 7:
            results[0] += 1
        else:
            results[1] += 1
    return results

print(f'[Wins, Losses] = {monte_carlo_simulation()}')
```

Results maybe favourable to us, but was that by luck?

Now do it 1000 times
Takes some time

```python
#Part 3
results = np.zeros(1000)

for i in range(1000):
    results[i] = monte_carlo_simulation()[0]

print(results.max())
```

Let us **PLOT** it

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(results, bins=15)
plt.show()
```

```python
#Our win/loss
print(results.mean())
print((results.mean()*5))
print(1000- results.mean())
print(results.mean()*4.75)
print(1000 -  results.mean())
print(results.mean()/1000)
```
